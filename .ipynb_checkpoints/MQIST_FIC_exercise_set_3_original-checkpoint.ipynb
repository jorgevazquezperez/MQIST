- $\hat\boldn\cdot \boldsig \ket{\pm,\hat\boldn} = \pm \ket{\pm,\hat\boldn} $

<div class="alert alert-block">

Let
$$
\ket{+,\hat\boldn} = \cos\frac{\theta}{2}\ket{0} + e^{i\phi}\sin\frac{\theta}{2}\ket{1}
$$
and
$$
\ket{-,\hat\boldn} = -e^{-i\phi}\sin\frac{\theta}{2}\ket{0} + \cos\frac{\theta}{2}\ket{1}
$$
with $\hat{\bf n}=(\sin\theta\cos\phi,\sin\theta\sin\phi, \cos\theta)^t$.

We want to prove that
$$
\hat \boldn\cdot \boldsig \ket{+,\hat\boldn} = + \ket{+,\hat\boldn}
$$
and
$$
\hat\boldn\cdot \boldsig \ket{-,\hat\boldn} = - \ket{-,\hat\boldn}.
$$

To prove the first statement, we can calculate $\hat\boldn\cdot  \ket{+,\hat\boldn}$ step by step:

$$
\begin{aligned}
\hat\boldn\cdot \boldsig \ket{+,\hat\boldn} &= \hat\boldn\cdot \boldsig \left(\cos\frac{\theta}{2}\ket{0} + e^{i\phi}\sin\frac{\theta}{2}\ket{1}\right) \\
&= \cos\frac{\theta}{2}(\hat\boldn\cdot \boldsig)\ket{0} + e^{i\phi}\sin\frac{\theta}{2}(\hat\boldn\cdot \boldsig)\ket{1}.
\end{aligned}
$$

Now, we need to calculate $(\hat\boldn\cdot \boldsig)\ket{0}$ and $(\hat\boldn\cdot \boldsig)\ket{1}$. The Pauli matrices are given by $\boldsig = (\sigma_x, \sigma_y, \sigma_z)$, so we have:

$$
\begin{aligned}
(\hat\boldn\cdot \boldsig)\ket{0} &= (\sin\theta\cos\phi\sigma_x + \sin\theta\sin\phi\sigma_y + \cos\theta\sigma_z)\ket{0} \\
&= \sin\theta\cos\phi\ket{1} - i\sin\theta\sin\phi\ket{1} + \cos\theta\ket{0} \\
&= \sin\theta(\cos\phi - i\sin\phi)\ket{1} + \cos\theta\ket{0}.
\end{aligned}
$$

Similarly,

$$
\begin{aligned}
(\hat\boldn\cdot \boldsig)\ket{1} &= (\sin\theta\cos\phi\sigma_x + \sin\theta\sin\phi\sigma_y + \cos\theta\sigma_z)\ket{1} \\
&= \sin\theta\cos\phi\ket{0}  i\sin\theta\sin\phi\ket{0} - \cos\theta\ket{1}) \\
&= \sin\theta(\cos\phi + \sin\phi)\ket{0} - \cos\theta\ket{1}.
\end{aligned}
$$

Now, let's continue with the calculation of $\hat\boldn\cdot \boldsig \ket{+,\hat\boldn}$:

$$
\begin{aligned}
\hat\boldn\cdot \boldsig \ket{+,\hat\boldn} &= \cos\frac{\theta}{2}(\hat\boldn\cdot \boldsig)\ket{0} + e^{i\phi}\sin\frac{\theta}{2}(\hat\boldn\cdot \boldsig)\ket{1} \\
&= \cos\frac{\theta}{2}\left(\sin\theta(\cos\phi - i\sin\phi)\ket{1} + \cos\theta\ket{0}\right) + e^{i\phi}\sin\frac{\theta}{2}\left(\sin\theta(\cos\phi + i\sin\phi)\ket{0} - \cos\theta\ket{1}\right).
\end{aligned}
$$

Now, let's simplify this expression:

$$
\begin{aligned}
\hat\boldn\cdot \boldsig \ket{+,\hat\boldn} &= \cos\frac{\theta}{2}\sin\theta(\cos\phi - i\sin\phi)\ket{1} + \cos\frac{\theta}{2}\cos\theta\ket{0} + e^{i\phi}\sin\frac{\theta}{2}\sin\theta(\cos\phi + i\sin\phi)\ket{0} - e^{i\phi}\sin\frac{\theta}{2}\cos\theta\ket{1}.
\end{aligned}ligned}
$$

Now, let's compare this r$ult to \(\ket{+,\h$\boldn}\) and see if it matches:

$$
\begin{aligned}
\ket{+,\hat\boldn} &= \cos\frac{\theta}{2}\ket{0} + e^{i\phi}\sin\frac{\theta}{2}\ket{1}.
\end{aligned}
$$

We can see that the two expressions match, with coefficients that are the same. Therefore, we have shown that

$$
\hat\boldn\cdot \boldsig \ket{+,\hat\boldn} = + \ket{+,\hat\boldn}.
$$

Now, let's prove the second statement:

To prove the second statement, we can calculate \(\hat\boldn\cdot \boldsig \ket{-,\hat\boldn}\) step by step. The calculation is similar to the one we just did, so I'll provide the result directly:

$$
\begin{aligned}
\hat\boldn\cdot \boldsig \ket{-,\hat\boldn} &= -e^{-i\phi}\sin\frac{\theta}{2}\sin\theta(\cos\phi + \sin\phi)\ket{1} - e^{-i\phi}\sin\frac{\theta}{2}\cos\theta\ket{0} + \cos\frac{\theta}{2}\sin\theta(\cos\phi + \sin\phi)\ket{0} - \cos\frac{\theta}{2}\cos\theta\ket{1}.
\end{aligned}
$$

Now, let's compare this r$ult to \(\ket{-,\h$\boldn}\) and see if it matches:

$$
\begin{aligned}
\ket{-,\hat\boldn} &= -e^{-i\phi}\sin\frac{\theta}{2}\ket{0} + \cos\frac{\theta}{2}\ket{1}.
\end{aligned}
$$

We can see that the two expressions match, with coefficients that are the same. Therefore, we have shown that

$$
\hat\boldn\cdot \boldsig \ket{-,\hat\boldn} = - \ket{-,\hat\boldn}.
$$

So, we have successfully proven both statements:

$$
\hat\boldn\cdot \boldsig \ket{+,\hat\boldn} = + \ket{+,\hat\boldn}
$$

and

$$
\hat\boldn\cdot \boldsig \ket{-,\hat\boldn} = - \ket{-,\hat\boldn}.
$$
