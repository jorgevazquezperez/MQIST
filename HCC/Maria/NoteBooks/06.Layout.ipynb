# Pases del transpilador - layout

En este NoteBook vamos a profundizar en uno de las etapas del transpilador, la selecci√≥n del layout o mapeo. 

```python
from qiskit_ibm_provider import IBMProvider

#Elijo como backend el computador ibm_kyoto
provider=IBMProvider()
backend=provider.get_backend("ibm_kyoto")
```

```python
from qiskit import QuantumCircuit, transpile

#Creo un circuito de 3 qubits
qc=QuantumCircuit(3)
qc.h(0)
qc.cx(0,1)
qc.cx(0,2)
qc.draw()
```

```python
from qiskit.visualization import plot_circuit_layout
#lo transpilo con nivel de optimzaci√≥n 0 y 1 y veo con lo mapea sobre el hardware
tqc=transpile(qc,backend,optimization_level=0)
display(plot_circuit_layout(tqc,backend))
        
tqc=transpile(qc,backend,optimization_level=1)
display(plot_circuit_layout(tqc,backend))      
```
