# Ejemplo de pase de planificación

En este ejemplo vamos a utilizar un pequeño circuito para ver la diferencia entre una planificación ASAP y ALAP. 

```python
from qiskit import QuantumCircuit

#Contruimos el circuito y lo dibujamos
circuito=QuantumCircuit(2)
circuito.x(0)
circuito.x(0)
circuito.x(1)
circuito.measure_all()
circuito.draw()
```

```python
from qiskit_ibm_provider import IBMProvider

#Selecciono el backend a utilizar
provider=IBMProvider()
backend=provider.get_backend("ibm_kyoto")
```

```python
from qiskit import transpile

#Transpilo el circuito para ese backend con una planificacion ASAP
#Incluyo nivel de optimizacion 0 para que no simplifique las 2 X consecutivas
tcircuito=transpile(circuito,backend,optimization_level=0,scheduling_method="asap")

tcircuito.draw()
```

```python
from qiskit import transpile

#Transpilo el circuito para ese backend con una planificacion ALAP
#Incluyo nivel de optimizacion 0 para que no simplifique las 2 X consecutivas
tcircuito=transpile(circuito,backend,optimization_level=0,scheduling_method="alap")

tcircuito.draw()
```
